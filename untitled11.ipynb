```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from itertools import combinations
from matplotlib.colors import ListedColormap
cmap_background = ListedColormap(['#FFAAAA', '#AAAAFF', '#AAFFAA'])
cmap_points = ListedColormap(['#FF0000', '#0000FF', '#00FF00'])
```

## ГЕНЕРАЦИЯ ДАННЫХ

```python
# список для координат точек
x = []

# список для меток классов
y = []

# параметры для спиралей
t = 2
s = 0.7
m = 0.02

# радиус первой спирали
r1 = np.log(np.arange(0.02, 10, m) + 1) * 4

# угол для первой спирали
theta1 = r1 / t

# добавляем шум в углы
for i in range(4):
    theta1 += np.random.uniform(0, s, len(r1))

# создаём точки первой спирали
for i in range(len(r1)):
    a = [0, 0]  # точка из двух координат
    a[0] = r1[i] * np.cos(theta1[i])  # x координата
    a[1] = r1[i] * np.sin(theta1[i])  # y координата
    x.append(a)  # добавляем точку
    y.append(0)  # класс 0


# радиус второй спирали
r2 = np.log(np.arange(0.02, 10, m) + 1) * 4

# угол второй спирали со сдвигом
theta2 = r2 / t + np.pi * 2 / 3

# добавляем шум
for i in range(4):
    theta2 += np.random.uniform(0, s, len(r2))

# создаём точки второй спирали
for i in range(len(r1)):
    a = [0, 0]
    a[0] = r2[i] * np.cos(theta2[i])
    a[1] = r2[i] * np.sin(theta2[i])
    x.append(a)
    y.append(1)  # класс 1


# радиус третьей спирали
r3 = np.log(np.arange(0.02, 10, m) + 1) * 4

# угол третьей спирали со сдвигом
theta3 = r3 / t + np.pi * 4 / 3

# добавляем шум
for i in range(4):
    theta3 += np.random.uniform(0, s, len(r3))

# создаём точки третьей спирали
for i in range(len(r1)):
    a = [0, 0]
    a[0] = r3[i] * np.cos(theta3[i])
    a[1] = r3[i] * np.sin(theta3[i])
    x.append(a)
    y.append(2)  # класс 2


# переводим списки в numpy массивы
x = np.array(x)
y = np.array(y)

# выводим данные
print("x:", x)
print()
print("y:", y)
print()


# создаём сетку для графика
xx, yy = np.meshgrid(
    np.arange(x[:, 0].min() - 1, x[:, 0].max() + 1, 0.01),
    np.arange(x[:, 1].min() - 1, x[:, 1].max() + 1, 0.01)
)

# рисуем точки
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=cmap_points, edgecolor='k')
```

## СОБСТВЕННАЯ РЕАЛИЗАЦИЯ RANDOM FOREST

```python
class RandomForestClassifier:
    # конструктор класса
    def __init__(self, num_of_trees=20, max_depth=25):
        self.forest = []  # список деревьев
        for i in range(num_of_trees):
            # создаём дерево решений и добавляем в лес
            self.forest.append(
                tree.DecisionTreeClassifier(
                    max_depth=max_depth,
                    random_state=random.randint(1, 42)
                )
            )

    # функция случайного перемешивания данных
    def __my_shuffle(self, X, y):
        t = random.sample(list(zip(X, y)), len(y))  # перемешиваем пары X и y
        n = random.randint(15000, 17000)  # случайный размер выборки
        X = np.array([elem[0] for elem in t])[:n]  # берём часть X
        y = np.array([elem[1] for elem in t])[:n]  # берём часть y
        return X, y

    # обучение леса
    def fit(self, X_train, y_train):
        for i in range(len(self.forest)):
            X, y = self.__my_shuffle(X_train, y_train)
            self.forest[i].fit(X, y)

    # предсказание
    def predict(self, X_test):
        preds = []

        # собираем предсказания всех деревьев
        for i in range(len(self.forest)):
            preds.append(self.forest[i].predict(X_test))

        preds = np.array(preds)

        result = []

        # голосование по большинству
        for i in range(preds.shape[1]):
            unique_classes, counts = np.unique(preds[:, i], return_counts=True)
            majority_class = unique_classes[np.argmax(counts)]
            result.append(majority_class)

        return np.array(result)
```

```python
# загружаем датасет с ценами домов
df = pd.read_csv("/content/sample_data/california_housing_train.csv")
df
```

```python
# признаки
X = df.iloc[:, :-1].values

# целевая переменная
y = df.iloc[:, -1].values
```

```python
# создаём случайный лес
rfc = RandomForestClassifier(num_of_trees=20, max_depth=25)

# обучаем модель
rfc.fit(X, y)
```

```python
# делаем предсказания
preds = rfc.predict(X)

# считаем точность
accuracy = accuracy_score(y, preds)
print(f'Accuracy: {accuracy}')
```

```python
# строим график реальных и предсказанных значений
plt.scatter(y, preds, alpha=0.01)
plt.xlabel('True Labels')
plt.ylabel('Predicted Labels')
plt.show()
```
